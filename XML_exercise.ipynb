# XML exercise

Using data from [**mondial database**](https://drive.google.com/file/d/14lFT4nWHgwN36ij4XZh6OUuup-K9qLgR/view?usp=sharing) find the answers to following questions:

1. 10 countries with the lowest infant mortality rates
2. 10 cities with the largest population
3. name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
import xml.etree.ElementTree as ET
tree = ET.parse("C:/Users/cwh93/Downloads/mondial.xml")
```

```python
root = tree.getroot()
root
```

```python
print(root.tag)
print(root.attrib)
print(len(root))
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
L={'name':[],'infant_mortality': []}
for country in root.findall('country'):   
    name = country.findtext('name')
    IM = country.findtext('infant_mortality')
    try:
        IM=float(IM)
    except:
        pass
    L['name'].append(name)
    L['infant_mortality'].append(IM)
df = pd.DataFrame(L) 
df.dropna().sort_values(by='infant_mortality',ascending=True).head(10)
```

```python
T={'Cname':[],'population': []}
for country in root.findall('country'):
    for city in country.findall('city'):
        Cname = city.findtext('name')
        POP=city.findtext('population')
        try:
            POP=float(POP)
        except:
            pass
        T['Cname'].append(Cname)
        T['population'].append(POP)
R= pd.DataFrame(T)
R.sort_values(by='population',ascending=False).head(10)
```

```python
longest_river = {'country':[],'river_name':[], 'length': [] }
for river in root.findall('river'):
        river_name = river.findtext('name')
        length = river.findtext('length')
        source = river.get('country')
        try:
            length = float(length)
        except:
            pass
        longest_river['river_name'].append(river_name)
        longest_river['length'].append(length)
        longest_river['country'].append(source)
longest_river = pd.DataFrame(longest_river)
longest_river.dropna().sort_values(by = 'length', ascending = False).head(1)
```

```python
longest_river = {'country':[],'lake_name':[], 'area': [] }
for lake in root.findall('lake'):
        river_name = lake.findtext('name')
        length = lake.findtext('area')
        source = lake.get('country')
        try:
            length = float(length)
        except:
            pass
        longest_river['lake_name'].append(river_name)
        longest_river['area'].append(length)
        longest_river['country'].append(source)
longest_river = pd.DataFrame(longest_river)
longest_river.dropna().sort_values(by = 'area', ascending = False).head(1)
```

```python
airport = {'country':[],'airport_name':[], 'elevation': [] }
for airports in root.findall('airport'):
        airport_name = airports.findtext('name')
        elevation = airports.findtext('elevation')
        source = airports.get('country')
        try:
            elevation = float(elevation)
        except:
            pass
        airport['airport_name'].append(airport_name)
        airport['elevation'].append(elevation)
        airport['country'].append(source)
airport = pd.DataFrame(airport)
airport.dropna().sort_values(by = 'elevation', ascending = False).head(1)
```
